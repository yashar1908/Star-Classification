```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('star_data.csv')
```

```python
df.head()
```

```python
X = df.iloc[:, :-3].join(df.iloc[:,-2:])
```

```python
X.head()
```

```python
Y = df.iloc[:,-3]
Y.head()
```

```python
X.describe()
```

```python
X.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X_train['Star color enc'] = le.fit_transform(X_train['Star color'])
```

```python
X_train['Spectral Class enc'] = le.fit_transform(X_train['Spectral Class'])
```

```python
X_train = X_train.drop('Star color', axis = 1)
```

```python
X_train = X_train.drop('Spectral Class', axis = 1)
```

```python
X_train.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn_model = KNeighborsClassifier(n_neighbors=6)
knn_model.fit(X_train, y_train)
```

```python
knn_model_2 = KNeighborsClassifier(n_neighbors=5)
knn_model_2.fit(X_train, y_train)
```

```python
X_test['Star color enc'] = le.fit_transform(X_test['Star color'])
X_test['Spectral Class enc'] = le.fit_transform(X_test['Spectral Class'])
X_test = X_test.drop('Star color', axis = 1)
X_test = X_test.drop('Spectral Class', axis = 1)
```

```python
y_pred = knn_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
accuracy
```

```python
y_pred_2 = knn_model_2.predict(X_test)
```

```python
accuracy_2 = accuracy_score(y_test, y_pred_2)
```

```python
accuracy_2
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=6, random_state=1)
rf_classifier.fit(X_train, y_train)
```

```python
y_pred_rf = rf_classifier.predict(X_test)
```

```python
accuracy_rf = accuracy_score(y_test, y_pred_rf)
```

```python
accuracy_rf
```

```python
from sklearn.metrics import classification_report, confusion_matrix
report_rf = classification_report(y_test, y_pred_rf)
```

```python
print(report_rf)
```

```python
cf = confusion_matrix(y_test, y_pred_rf)
print(cf)
```
